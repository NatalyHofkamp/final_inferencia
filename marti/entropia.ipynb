## Entropía vista en dados

```python
import random 
from math import *
import numpy as np
from matplotlib import pyplot as plt
```

```python

def H (p):
    return -p*np.log2(p)-(1-p)*log2(1-p)

Hs = []
ps = np.linspace(0,1,1000)
for p in ps:
    if p != 0 and p != 1:
        Hs.append(H(p))
    else: 
        Hs.append(0)

plt.plot(ps,Hs)
plt.show()
```

```python
def marg(x, anch):
    """Cálculo de marginal. """

    p_x, _ = np.histogram(x, anch)
    p_x = p_x/len(x)
    return p_x


def h(x, anch):
    """Cálculo de entropía de una variable. """
    p_x = marg(x, anch)
    
    h = 0
    for p in p_x:
        h -= p*log2(p)

    return h
```
