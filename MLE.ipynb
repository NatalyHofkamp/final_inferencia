1. Densidad Exponencial

Sea $x$ una variable aleatoria con una densidad exponencial dada por:

\begin{equation*}

p(x|\theta) =
\begin{cases}
    \theta e^{-\theta x}, & \text{si } x \geq 0, \\
    0, & \text{en otro caso}.
\end{cases}
\end{equation*}


(a) Gráfico de $p(x|\theta)$ versus $x$ para $\theta = 1$. Gráfico de $p(x|\theta)$ versus $\theta$, $0<\theta<5$ , para $x = 2$



```python
lamb = 1
expos = np.arange(0, 10, 0.1)

def p_expo(x, lam):
    proba = 0
    if x >= 0:
        proba = lam*np.exp(-lam*x)
    return proba

plt.figure()
plt.scatter(expos, np.array([p_expo(x, lamb) for x in expos]))
plt.title('')
plt.xlabel('x')
plt.ylabel('p(x | lambda=1)')
plt.show()
plt.close()
```

```python
x = 2
lambs = np.arange(0, 5, 0.1)

def p_expo(x, lam):
    proba = 0
    if x >= 0:
        proba = lam*np.exp(-lam*x)
    return proba

plt.figure()
plt.scatter(lambs, np.array([p_expo(2, l) for l in lambs]))
plt.xlabel('lambda')
plt.ylabel('p(x=2 | lambda)')
plt.show()
plt.close()
```

Let $x$ have a uniform density

\begin{equation*}
p(x|\theta) =
\begin{cases}
    \frac{1}{\theta}, & \text{si } 0 \leq x \leq \theta, \\
    0, & \text{en otro caso}.
\end{cases}
\end{equation*}

$p(x|\theta) \sim U(0, \theta)$.

(a) Supongamos que $n$ muestras $D = \{x_1, \ldots, x_n\}$ son obtenidas independientemente de acuerdo con $p(x|\theta)$. Muestra que la estimación de máxima verosimilitud para $\theta$ es $\max[D]$, es decir, el valor del elemento máximo en $D$.


- - - - - - resolución - - - - - - 

L() = $p(x_1|\theta)$ . . .  $p(x_n|\theta)$
    = $\frac{1}{\theta} . . .  \frac{1}{\theta}$
    = $ \frac{1}{\theta^n}$

derivo e igualo a cero

$ \frac{-n}{\theta^(n+1)} = 0 $

 y esto se cumple cuando $n→+inf$

2.b) Suppose that $n = 5$ points are drawn from the distribution and the maximum value of which happens to be $\max x_k = 0.6$.

Plot the likelihood $p(D|\theta)$ in the range $0 \leq \theta \leq 1$.
Explain in words why you do not need to know the values of the other four points.


```python
unif_rand = np.arange(0, 1, 0.1)
unif_max = 0.6
def p_unif(unif_max, x):
    tita = 1/unif_max
    if x < 0 or x > 0.6:
        return 0
    return tita

plt.figure()
plt.scatter(unif_rand, np.array([p_unif(unif_max, u) for u in unif_rand]))
plt.title('')
plt.xlabel('x')
plt.ylabel('p(x | tita)')
plt.show()
plt.close()
```
